```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

numerical_columns = ['price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'calculated_host_listings_count', 'availability_365']

scaler = StandardScaler()
df[numerical_columns] = scaler.fit_transform(df[numerical_columns])
df
```

```python
from sklearn.preprocessing import RobustScaler
numerical_columns = ['price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'calculated_host_listings_count', 'availability_365']
scaler = RobustScaler()
df[numerical_columns] = scaler.fit_transform(df[numerical_columns])

# Function to detect outliers using the IQR method
def detect_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return ((df[column] < lower_bound) | (df[column] > upper_bound)).sum()

# Check for outliers in each numerical column after scaling
outliers_count = 0
for col in numerical_columns:
    num_outliers = detect_outliers_iqr(df, col)
    print(f"Outliers in {col}: {num_outliers}")
    outliers_count += num_outliers

print(f"Total number of outliers across all numerical columns: {outliers_count}")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
sns.histplot(df['price'], kde=True)
plt.title('Original Data')
df['price_log'] = np.log(df['price'] + 1)
plt.subplot(1, 2, 2)
sns.histplot(df['price_log'], kde=True)
plt.title('Log Transformed Data')
plt.show()
print("Original Skewness: ", df['price'].skew())
print("Log Transformed Skewness: ", df['price_log'].skew())
```
