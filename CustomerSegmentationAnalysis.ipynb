```python
Q1 = df['Income'].quantile(0.25)
Q3 = df['Income'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = df[(df['Income'] < lower_bound) | (df['Income'] > upper_bound)]

print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)
print("Number of Outliers:", len(outliers))
```

```python
Q1 = df['MntTotal'].quantile(0.25)
Q3 = df['MntTotal'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = df[(df['MntTotal'] < lower_bound) | (df['MntTotal'] > upper_bound)]

print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)
print("Number of Outliers:", len(outliers))
```

```python
# Filter out outliers
df_no_outliers = df[(df['MntTotal'] >= lower_bound) & (df['MntTotal'] <= upper_bound)]

# Print the number of rows before and after removing outliers
print("Number of Rows Before Removing Outliers:", len(df))
print("Number of Rows After Removing Outliers:", len(df_no_outliers))
```

```python
def get_marital_status(row):
    if row['marital_Divorced'] == 1:
        return 'Divorced'
    elif row['marital_Married'] == 1:
        return 'Married'
    elif row['marital_Single'] == 1:
        return 'Single'
    elif row['marital_Together'] == 1:
        return 'Together'
    elif row['marital_Widow'] == 1:
        return 'Widow'
    else:
        return 'Unknown'

# Apply the function to create a new 'Marital' column
df['Marital'] = df.apply(get_marital_status, axis=1)

# Create a bar plot
plt.figure(figsize=(8, 6))
sns.barplot(x='Marital', y='MntTotal', data=df, palette='viridis')
plt.title('MntTotal by marital status')
plt.xlabel('Marital status')
plt.ylabel('MntTotal')
plt.show()
```

```python
def get_educational_status(row):
    if row['education_2n Cycle'] == 1:
        return '2n Cycle'
    elif row['education_Basic'] == 1:
        return 'Basic'
    elif row['education_Graduation'] == 1:
        return 'Graduation'
    elif row['education_Master'] == 1:
        return 'Master'
    elif row['education_PhD'] == 1:
        return 'PhD'
    else:
        return 'Unknown'

# Apply the function to create a new 'Education' column
df['Education'] = df.apply(get_educational_status, axis=1)

# Create a bar plot
plt.figure(figsize=(8, 6))
sns.barplot(x='Education', y='MntTotal', data=df, palette='viridis')
plt.title('MntTotal by educational status')
plt.xlabel('Educational status')
plt.ylabel('MntTotal')
plt.show()
```

```python
def get_products_status(row):
    if row['MntWines'] == 1:
        return 'Wines'
    elif row['MntFruits'] == 1:
        return 'Fruits'
    elif row['education_Graduation'] == 1:
        return 'Graduation'
    elif row['MntMeatProducts'] == 1:
        return 'MeatProducts'
    elif row['MntFishProducts'] == 1:
        return 'FishProducts'
    elif row['MntSweetProducts'] == 1:
        return 'SweetProducts'
    elif row['MntGoldProds'] == 1:
        return 'GoldProds'
    else:
        return 'Unknown'

# Apply the function to create a new 'Education' column
df['Products'] = df.apply(get_products_status, axis=1)

# Create a bar plot
plt.figure(figsize=(8, 6))
sns.barplot(x='Products', y='MntTotal', data=df, palette='viridis')
plt.title('MntTotal by educational status')
plt.xlabel('Products status')
plt.ylabel('MntTotal')
plt.show()

```

```python
from sklearn.cluster import KMeans

X = df[['MntTotal', 'Income']]

# Specify the number of clusters
n_clusters = 3

# Instantiate the KMeans model
kmeans = KMeans(n_clusters=n_clusters, random_state=42)

# Fit the model to the data
kmeans.fit(X)

# Get cluster labels
cluster_labels = kmeans.labels_

# Add cluster labels to the original DataFrame
df['Cluster'] = cluster_labels

# Print the counts of each cluster
print(df['Cluster'].value_counts())

# Print the centroids of each cluster
print(kmeans.cluster_centers_)
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
X = df[['MntTotal', 'Income','MntRegularProds']]
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
df_pca = pd.DataFrame(data=X_pca, columns=['PC1', 'PC2'])
print("Explained Variance Ratio:", pca.explained_variance_ratio_)
print("Cumulative Explained Variance Ratio:", pca.explained_variance_ratio_.cumsum())
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
inertia = []
k_values = range(1, 11)
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(10, 6))
plt.plot(k_values, inertia, marker='o', linestyle='-')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.xticks(k_values)
plt.grid(True)
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)
cluster_labels = kmeans.labels_
df['Cluster'] = cluster_labels
print(df['Cluster'].value_counts())
print(kmeans.cluster_centers_)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(X.iloc[:, 0], X.iloc[:, 1], c=cluster_labels, cmap='viridis', marker='o', alpha=0.5, label='Data Points')

# Plot the centroids
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', marker='X', s=100, label='Centroids')

# Add labels and legend
plt.title('K-means Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid(True)
plt.show()
```

```python
cluster_means = df.groupby('Cluster')['MntTotal'].mean()

# Plot the bar plot
plt.figure(figsize=(10, 6))
plt.bar(cluster_means.index, cluster_means.values, color='skyblue')
plt.title('Mean MntTotal by Cluster')
plt.xlabel('Cluster')
plt.ylabel('Mean MntTotal')
plt.xticks(cluster_means.index)
plt.grid(axis='y')
plt.show()
```
